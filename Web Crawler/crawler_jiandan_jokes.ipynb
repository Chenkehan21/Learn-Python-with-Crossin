```python
import requests
import re
from lxml import etree
```

```python
url = "http://jandan.net/treehole"
req = requests.get(url)
req.encoding = 'utf-8'
print(req.status_code)
print(req.text)
print(type(req.text))
```

```python
# survey the web page resource, we find that the text we need located between <p></p>
info1 = re.findall(r'<p>.*</p>', req.text)
for text in info1:
    print(text)
    print('='*100)
# another way to extract text
# pattern = re.compile(r'<p>.*</p>')
# info1 = pattern.findall(req.text)
# info1
```

```python
# There are still many problems:
# some text has been newlined, we failed to extract them.
# we just need the text i.e. we should drop other characters.
# new_pattern = re.compile(r'</a></span><p>([\s\S]*?)</p>\n                        </div>')
new_pattern = re.compile(r'</a></span><p>([\s\S]*?)</p>\n.*</div>') # use '.*' to represent the long space !!!
# survey more carefuly, we found that the text we need is between </a></span><p> and </div>
# in order to extract text that was newlined, we should use [\s\S]*?
# in order to drop those useless charasters, we add a parenthesis.
# finally we replace things like <p>; </p>; </ br> with ''
info2 = new_pattern.findall(req.text)
# print(info2)
for text in info2:
    text = re.sub(r'[<p></p><br />]', '', text)
    print(text)
    print('='*100)
```

```python
# METHOD 2: USE LXML !!!
tree = etree.HTML(req.text)
# print(tree, '\n')
# look all the "li" nodes for the "div" with class = text
result = tree.xpath('//li//div[@class="text"]')
# print(result,'\n')
for div in result:
#     print(etree.tostring(div), '\n')
    content = div.xpath('p/text()') # there's no '/' before 'p'. it means choose from the child nodes.
#     print(content)
    for p in content:
        print(p)
    print('='*100)
```
