```python
import pandas as pd
from pyecharts.charts import Bar
from pyecharts.charts import Pie
from pyecharts import options as opts
from snapshot_selenium import snapshot
from pyecharts.render import make_snapshot
```

# 读取 Task_4_lianjia_sale.csv 数据(阅读参考文章 2、3)，利用 groupby 函数，以【区域】字段进行 groupby， 同时结合 agg 聚合函数统计各个区域平均房价; 将统计所得平均房价命名为【区域平均价格】列 (阅读参考文章 4、5)并将数据从小到大按顺序排列(提示:df.sort_values(by = “ ”) (阅读参考文章 6));最后展示所得前 5 行数据 (将所得的 DataFrame 命名为 df_avg);

```python
with open("data/Task_4_lianjia_sale.csv", encoding='utf-8') as f:
    lianjia_sale = pd.read_csv(f)
lianjia_sale.head(10)
```

```python
df_avg = lianjia_sale.groupby(['区域']).价格.agg([('区域平均价格', 'mean')])
formater="{0:.03f}".format
pd.set_option('precision', 2) # control decimal places!!!!!
df_avg.sort_values(by=['区域平均价格']).head()
```

# 利用 Pyecharts 将第(1)问所得数据作条形图(阅读参考文章 7)，图表标题为“上海各区域平均价格”(提示: 最后用 render_notebook()渲染显示出图片);

```python
locations = list(df_avg.index)
avg_price = [round(x, 2) for x in df_avg['区域平均价格'].values.tolist()] 
# use round() to control the values' decimal places !!!
```

```python
bar = (
    Bar()
    .add_xaxis(locations)
    .add_yaxis("price", avg_price)
    .set_global_opts(title_opts={"text": "上海各区域平均价格"})
)
bar.render_notebook()
```

```python
make_snapshot(snapshot, bar.render(), "上海各区域平均价格(作业4).png")
```

# 利用 groupby 函数，以【装修类型】字段进行 groupby，同时结合 agg 聚合函数统计各个区域房源数量; 将统计所得数据命名为【房源数量】列;最后展示所得数据(将所得的 DataFrame 命名为 df_zhuangxiu);

```python
df_zhuangxiu = lianjia_sale.groupby(['装修类型', '区域']).小区.agg([('房源数量', 'count')])
df_zhuangxiu
```

```python
def get_type(decoration_type):
    info = lianjia_sale[lianjia_sale['装修类型'] == ' %s ' % decoration_type]
    df_info = info.groupby('区域').小区.agg([('房源数量', 'count')])
    return df_info
```

```python
df_type_others = get_type("其他")
df_type_others
```

```python
df_type_rough = get_type('毛坯')
df_type_rough
```

```python
df_type_simple = get_type('简装')
df_type_simple
```

```python
df_type_exquisite = get_type('精装')
df_type_exquisite
```

# 利用 Pyecharts 将第(3)问所得数据作饼图(阅读参考文章 7、8)，图表标题为“装修类型占比情况”(提示: v1=["啤酒","可乐","雪碧","咖啡","奶茶"] v2=[30,19,21,12,18]，a = [list(z) for z in zip(v1,v2)]， 则生成 v1 与 v2 一一对应的数据)。

```python
pie = (
    Pie()
    .add("", [list(x) for x in zip(list(df_zhuangxiu.index), df_zhuangxiu['房源数量'].values.tolist())])
    .set_global_opts(title_opts=opts.TitleOpts(title='装修类型占比情况'))
    .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}"))
)
pie.render_notebook()
```

```python
make_snapshot(snapshot, pie.render(), "装修类型占比情况(作业4).png")
```

```python
def draw_pie(df_name, decoration_type):
    pie = (
        Pie()
        .add("", [list(x) for x in zip(list(df_name.index), df_name['房源数量'].values.tolist())])
        .set_global_opts(title_opts=opts.TitleOpts(title='各区域装修类型为”%s“的占比情况'%decoration_type, pos_bottom=True))
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}:{c}"))
    )
    make_snapshot(snapshot, pie.render(), "各区域装修类型为”%s“的占比情况(作业4).png"%decoration_type)
    return pie.render_notebook()
```

```python
draw_pie(df_type_others, "其他")
```

```python
draw_pie(df_type_rough, '毛坯')
```

```python
draw_pie(df_type_simple, "简装")
```

```python
draw_pie(df_type_exquisite, "精装")
```
