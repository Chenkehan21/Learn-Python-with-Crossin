# 根据任务 4 所统计的上海各个区的房均价数据，利用 Pyecharts 中的 Map，以上海地图为背景作可视化展示图 (提示:可以直接定义 districts = ['金山区', '奉贤区', '嘉定区', '松江区', '青浦区', '宝山区', '闵行区 ', '浦东新区', '普陀区', '杨浦区', '虹口区', '长宁区', '静安区', '徐汇区', '黄浦区'])(阅读参考文章 2);

```python
with open("data/Task_4_lianjia_sale.csv", encoding='utf-8') as f:
    lianjia_sale = pd.read_csv(f)
lianjia_sale.head(10)
```

```python
avg_price = lianjia_sale.groupby('区域').价格.agg({'平均房价': 'mean'}).sort_values(by='平均房价')
avg_price
```

```python
location1 = [i + '区' for i in list(avg_price.index)]
location2 = ['浦东新区' if x == '浦东区' else x for x in location1]
map_shanghai = (
    Map()
    .add("区域平均房价", [list(x) for x in zip(location2, avg_price['平均房价'].values.tolist())], "上海")
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Map-上海各区域平均房价"),
        visualmap_opts=opts.VisualMapOpts(max_=max(avg_price['平均房价'].values.tolist())),
    )
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
)
map_shanghai.render_notebook()
```

```python
make_snapshot(snapshot, map_shanghai.render(), "上海各区域平均房价.png")
```

# 导入 tushare 数据包，提取 A 股基本数据信息，利用 groupby 函数统计每个省份的公司数量，并以中国地图 为背景作相应图表(阅读参考文章 2、3、4)

```python
df
```

```python
area_industry_num = df.groupby('area').name.agg(['count'])
area_industry_num
```

```python
area = ["内蒙古" if x == '内蒙' else x for x in list(area_industry_num.index)]
map_china = (
    Map()
    .add("公司数量",
         [list(x) for x in zip(area, area_industry_num['count'].values.tolist())],
         "china"
        )
    .set_global_opts(
        title_opts = opts.TitleOpts(title="Map-各省份公司数量"),
        visualmap_opts = opts.VisualMapOpts(max_ = max(area_industry_num['count'].values.tolist())),
    )
)
map_china.render_notebook()
```

```python
make_snapshot(snapshot, map_china.render(), "中国个省份公司数量.png")
```
