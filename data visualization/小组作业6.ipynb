```python
import pandas as pd
import tushare as ts
from pyecharts import options as opts
from pyecharts.charts import Liquid, Page
from pyecharts.charts import Gauge, Page
from snapshot_selenium import snapshot
from pyecharts.render import make_snapshot
from pyecharts.charts import WordCloud
from pyecharts.globals import SymbolType
from pyecharts.charts import Line
```

# 导入 tushare 数据包，提取 A 股基本数据信息，取毛利率 gpr 和净利率 npr 的平均数，并分别以水球图和仪 表盘图表现出来(阅读参考文章 2、3);

```python
df = ts.get_stock_basics()
df.head()
```

```python
gpr = (
    Liquid()
    .add("", [df['gpr'].values.mean()/100])
    .set_global_opts(title_opts=opts.TitleOpts(title="GPR"))
)
gpr.render_notebook()
```

```python
make_snapshot(snapshot, gpr.render(), "gpr(作业6).png")
```

## make_snapshot(snapshot, gpr.render(), "gpr(作业6).png")

```python
npr = (
    Gauge()
    .add('', [("净利率", round(df['npr'].values.mean(), 2))])
    .set_global_opts(title_opts=opts.TitleOpts("NPR"))
)
npr.render_notebook()
```

```python
make_snapshot(snapshot, npr.render(), "npr(作业6).png")
```

# 统计 A 股各行业公司数量，并根据行业分布数据作词云图(阅读参考文章 4、5);

```python
industry_num = df.groupby('industry').name.agg([('number', 'count')]) # give the column a name!
industry_num
```

```python
words = [x for x in zip(list(industry_num.index), industry_num['count'].values.tolist())]
words
```

```python
industry = (
    WordCloud()
    .add("", words, word_size_range=[10, 100], shape=SymbolType.DIAMOND)
    .set_global_opts(title_opts=opts.TitleOpts("行业分布"))
)
industry.render_notebook()
```

```python
make_snapshot(snapshot, industry.render(), "行业分布(作业6).png")
```

# 利用 tushare 提取茅台数据中的最高最低价，作双折线图，茅台股票代码 600519 (提示:提取单个股票信息 df = ts.get_hist_data('600519',start=' ',end=' ') )(阅读参考文章 1);

```python
maotai = ts.get_hist_data("600519")
maotai.head()
```

```python
maotai.index
```

```python
high = list(maotai['high'])
low = list(maotai['low'])
```

```python
high_low = (
    Line()
    .add_xaxis(list(maotai.index))
    .add_yaxis("high", high)
    .add_yaxis("low", low)
    .set_global_opts(title_opts=opts.TitleOpts("High vs Low"))
)
high_low.render_notebook()
```

```python
make_snapshot(snapshot, high_low.render(), "High vs Low(作业6).png")
```
