```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# 读取 CSV 格式文件“task_2_lianjia_data.csv”，将数据命名为 lianjia_data，并选取前部数据查看(阅读 参考文章 1、2);

```python
with open("task_2_lianjia_data.csv", encoding='utf-8') as f:
    lianjia_data = pd.read_csv(f)
lianjia_data.head()
```

# 利用 groupby 函数，以【面积大小(m²)】字段进行 groupby，统计各个面积下的【房源数量】情况(阅读参考 文章 3、4);

```python
info = lianjia_data.groupby('面积大小（㎡）').朝向.agg(['count'])
info.head()
```

# 将第(2)问所统计数据，作折线图和条形图，并得出相关结论(阅读参考文章 5、6)，提示:先提取第(2)问 所得数据表的 index 索引，并转化成 list;再提取【房源数量】的值转成 list(阅读参考文章 7、8);

```python
square = list(info.index)
num = info['count'].values.tolist()
print("square: ", square)
print('\nnum: ', num)
```

```python
plt.figure(figsize=(12, 10))
plt.plot(square, num, linewidth = 2.5, color = 'blue')
plt.xticks(ticks=np.arange(0, 175, 10), fontsize = 15)
plt.yticks(ticks=np.arange(0, 40, 5), fontsize = 15)
plt.xlabel('\nsquare(m²)', fontsize=16)
plt.ylabel('numbers\n', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(50, 12))
plt.bar(square, num, width = 0.8, facecolor = 'green')
plt.xticks(ticks=np.arange(0, 175, 5), fontsize = 30)
plt.yticks(ticks=np.arange(0, 40, 5), fontsize = 30)
plt.xlabel('\nsquare(m²)', fontsize=40)
plt.ylabel('numbers\n', fontsize=40)
plt.show()
```

# 利用 groupby 函数，以【区】字段进行 groupby，统计各个上海各个“区”所属的【房源数量】情况;

```python
info2 = lianjia_data.groupby('区').朝向.agg(['count'])
info2.head()
```

```python
block = info2.index.to_list()
num2 = info2['count'].values.tolist()
print(block, type(block), len(block))
print(num2)
```

# 将第(4)问所统计数据，作条形图，并得出相关结论;

```python
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize = (12, 10))
plt.bar(range(13), num2)
plt.xticks(ticks=range(13),
           labels = ['嘉定', '奉贤', '徐汇', '普陀', '杨浦', '松江', '浦东', '虹口', '长宁', '闵行', '青浦', '静安', '黄浦'],
          fontsize = 15)
plt.yticks(ticks = np.arange(0, 140, 10), fontsize = 15)
for a, b in zip(range(13), num2):    
    plt.text(a, b, '%d' % b, ha ='center', va='bottom', fontsize=20)
plt.xlabel('\n街区', fontsize = 25, color = 'green')
plt.ylabel('数量\n', fontsize = 25, color = 'green')
plt.show()
# 结论：徐汇区的房源最多，为126，奉贤区的房源最少，为1。
# 徐汇区、浦东区和静安区的房源都上百，嘉定区、奉贤区、普陀区、松江区、虹口区和青浦区的房源在50以下。
# 房源的地域分布不是很均匀。
```

# 利用 groupby 函数，以【楼层类型】字段进行 groupby，统计各个楼层所属的【房源数量】情况;

```python
info3 = lianjia_data.groupby('楼层类型').朝向.agg(['count'])
info3
```

# 将第(6)问所统计数据，作饼图(阅读参考文章 9)，并得出相关结论。

```python
species = [x.strip() for x in list(info3.index)]
num3 = info3['count'].values.tolist()
print(species)
print(num3)
```

```python
plt.figure(figsize=(10, 8))
plt,pie(num3, labels=species, autopct="%.2f%%", textprops={'fontsize': 28})
plt.show()

# 总结： 高楼层几乎占据了所有房源的一般；地下室的房源最少；中楼层比低楼层的房源更多。
```
