```python
import random

print("You need to guess the number I think." + '\n' + "hint: It's an integer from 1 to 50")
print('\n' + "Round1")

def guess_and_check():
    global guess_number
    global Bingo
    num_type = False
    while num_type == False:
        try:
            guess_number = int(input("Please enter your answer: "))
        except:    
            print("You must enter an integer!!!")
        else:
            if type(guess_number) == int:
                Bingo = False
                break

def compare(result, count):
    guess_and_check()
    if guess_number < result:
        print("Too small!")
        return False
    elif guess_number > result:
        print("Too big!")
        return False
    else:
        print("Bingo!!!")
        print("You guessed %d times" % count + '\n')
        num.append(count)
        return replay()  # there must be "return replay()"!!! if omit "return", it doesn't work!!!
        
def replay():
    global Round
    global count
    global result
    restart = input("Do you want to replay? (Y/N): ")
    if restart == "Y":
        result = random.randint(1, 50)
        Round += 1
        print('\n' + "Round%d" % Round)
        count = 0
        return False
    else:
        average = sum(num) / len(num)
        print('\n' + "You played % rounds, " % Round 
                + "You need to guess %.2f times per round" % average)
        return True

def play():
    global result
    result = random.randint(1, 10)
    global count
    while not compare(result, count):
        count += 1

count = 1
Round = 1
num = []
play()
```
